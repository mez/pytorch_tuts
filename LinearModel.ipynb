# This is a simple Linear Model with PyTorch

```python
import torch
from torch.autograd import Variable
```

```python
x_data = Variable(torch.Tensor([[1.0], [2.0], [3.0]]))
y_data = Variable(torch.Tensor([[2.0], [4.0], [6.0]]))
```

### Step 1: Define the Model

```python
class SimpleLinearModel(torch.nn.Module):
    def __init__(self):
        super(SimpleLinearModel, self).__init__();
        self.linear = torch.nn.Linear(1,1)
    
    def forward(self, x):
        '''
        The forward method accepts a Variable and must return a Variable.
        The idea is to use the model you've created and run the input through 
        the model, returning the output.
        '''
        y_pred = self.linear(x)
        return y_pred
```

```python
model = SimpleLinearModel()
```

### Step 2: Construct the loss and optimizer

```python
criterion = torch.nn.MSELoss(size_average=False)
optimizer = torch.optim.SGD(model.parameters(), lr=1e-2)
```

### Step 3: Training - forward, loss, backward, step

```python
for epoch in range(500):
    #foward pass
    y_pred = model(x_data)
    
    #compute loss
    loss = criterion(y_pred, y_data)
    print(epoch, loss.data[0])
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

### Step 4: Testing the Model

```python
datapoint = Variable(torch.Tensor([[4.0]])) #y_pred should be 8
print("Predicted (after training)", 4, model.forward(datapoint).data[0][0])
```
